### **3. Momentum Analysis - Each team performance in first-half and second-half of the match**

# **Recommendations**

**1. Raid-Dominant Teams** - Patna Pirates, U Mumba, Telugu Titans, Bengal Warriors, Bengaluru Bulls, Jaipur Pink Panthers need to strengthen defensive units to avoid over-reliance on raids. Use substitutions strategically to bring in defenders late in matches. Train raiders for do-or-die situations to maximize clutch scoring.

**2. Defense-Heavy Teams** - Storm Queens, Services SSCB, Fire birds need to train secondary raiders to reduce stagnation and over‑reliance on defense. Encourage aggressive raiding strategies to balance scoring. Focus on counter-attacks after successful tackles.

**3. Balanced Teams** - India, India Women, Services SCB teams need to maintain tactical balance but improve clutch raid efficiency in do‑or‑die situations. Use analytics to identify optimal raid timings (early vs late game). Leverage balanced squads to adapt against raid-heavy opponents.

**4. Fitness & Conditioning** - Teams with declining second-half performance should focus on stamina training and rotation strategies. Strong finishers should leverage momentum by pressing harder in final minutes.

**5. Analytics Application** - Continuously track raid vs tackle ratios per match using pivot tables. Monitor review outcomes to optimize challenge strategies. Use heatmaps to identify event clusters like when super raids or tackles occur most often.
